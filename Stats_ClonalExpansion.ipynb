# If not normally distributed use a Mann-Whitney test compare controls to cases. 
In this case normal controls are not normally distributed, so we'll run it all as non-parametric. 

```python
# Make the df x and y for the Mann-Whitney U test.

x = df_VAR_IsitNorm[COL]
y = df_CASE_IsitNorm[COL]
```

```python
# Try this Mann-Whitney test
# This is calculated on the proportion of the sample that is expanded to the level set in COL. 
# There is not weight attributed to sample size. 
# This is based on the medians, and does not return a confidence interval. There is no set standard for CI interval calculation in Mann Whitney testing. https://doi.org/10.1177/09622802188145.
# axis set to None or 0 return the same statistic. 

sp.stats.mannwhitneyu(x, y, alternative='two-sided', axis=0, method='exact')
```
